```python
#importing the necessary libraries
import requests
import bs4
```

```python
#creating an object of the class requests
response = requests.get('https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_area')
response.text
```

```python
soup = bs4.BeautifulSoup(response.text, "lxml")
```

```python
#Selecting the title
soup.select("title")
```

```python
#simple for loop for all p-tags text
for i in soup.select('p'):
    print(i.text)
```

```python
#using findall to get all links in the webpage
for link in soup.find_all('a', href=True):
    print(link['href'])
```
